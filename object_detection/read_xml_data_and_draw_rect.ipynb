```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import glob
```

```python
image=cv2.imread("D:\PASCAL\VOCdevkit\VOC2012\JPEGImages/2007_000027.jpg",cv2.IMREAD_COLOR)
```

```python
from xml.etree.ElementTree import parse,dump
tree = parse("D:/PASCAL/VOCdevkit/VOC2012/Annotations/2007_000027.xml")
# dump(tree)
```

```python
xmin = tree.findtext("object/bndbox/xmin")
ymin = tree.findtext("object/bndbox/ymin")
xmax = tree.findtext("object/bndbox/xmax")
ymax = tree.findtext("object/bndbox/ymax")
xmin=int(xmin)
ymin=int(ymin)
xmax=int(xmax)
ymax=int(ymax)

cv2.rectangle(image,(xmin, ymin),(xmax, ymax),(0,255,0),3)
print(xmin, ymin)
print(xmax, ymax)
type(xmin)
plt.imshow(image)
```

```python
image_data=[]
for i in glob.glob("D:\PASCAL\VOCdevkit\VOC2012\JPEGImages\*.jpg"):# load image data
    n=cv2.imread(i,cv2.IMREAD_COLOR)
    n=cv2.cvtColor(n, cv2.COLOR_BGR2RGB)# matplotlib = RGB channel , opencv = BGR channels
    image_data.append(n)
```

```python

xml_data=[]
for i in glob.glob("D:\PASCAL\VOCdevkit\VOC2012\Annotations\*.xml"):# load image data
    tree = parse(i)
    xml_data.append(tree)
```

```python
print(len(xml_data))
print(len(image_data))
```
