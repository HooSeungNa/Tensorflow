```python
def interaction_over_union(box1,box2):
    x1=max(box1[0],box2[0])
    y1=max(box1[1],box2[1])
    x2=min(box1[2],box2[2])
    y2=min(box1[3],box2[3])
    area_intersection = (x2 - x1) * (y2 - y1)#겹치는 곳의 넓이
    area_box1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area_box2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    area_union = area_box1 + area_box2 - area_intersection
    if(area_union!=0):
        iou = area_intersection/area_union
    else:
        iou=0
    return iou
```

```python
def indent(elem, level=0):
    i = "\n" + level*"  "
    if len(elem):
        if not elem.text or not elem.text.strip():
            elem.text = i + "  "
        if not elem.tail or not elem.tail.strip():
            elem.tail = i
        for elem in elem:
            indent(elem, level+1)
        if not elem.tail or not elem.tail.strip():
            elem.tail = i
    else:
        if level and (not elem.tail or not elem.tail.strip()):
            elem.tail = i
```

```python

import selectivesearch

for i in range(len(image_data)):#image 갯수 만큼 반복
    cnt_box=0
    image_num=Element("image_"+str(i))
    note.append(image_num)
    positive_box=[]
    negative_box=[]
    xmin = xml_data[i].findtext("object/bndbox/xmin")
    ymin = xml_data[i].findtext("object/bndbox/ymin")
    xmax = xml_data[i].findtext("object/bndbox/xmax")
    ymax = xml_data[i].findtext("object/bndbox/ymax")
    xmin=int(float(xmin))
    xmax=int(float(xmax))
    ymin=int(float(ymin))
    ymax=int(float(ymax))
    
    img = image_data[i]
    box1=[xmin,ymin,xmax,ymax]# ground truth
    # perform selective search
    img_lbl, regions = selectivesearch.selective_search(
        img, scale=500, sigma=0.8, min_size=20)
#     print(len(regions))
    candidates = set()
    for r in regions:
#         # excluding same rectangle (with different segments)
        if r['rect'] in candidates:
            continue
        # excluding regions smaller than 2000 pixels
        if r['size'] < 2000:
            continue
#         # distorted rects
        x, y, w, h = r['rect']
#         if w / h > 1.2 or h / w > 1.2:
#             continue
        candidates.add(r['rect'])
        cnt_box+=1
        
    
    for x, y, w, h in candidates:
#         print(x, y, w, h)
        box2=[x,y,x+w,x+h]#SS 를 통해서 생성된 관심영역
        resized_box=[0,0,0,0]
        resized_image=cv2.resize(img,(14,14))
        cur_iou=interaction_over_union(box1,box2)
        w_scale=np.round(resized_image.shape[0]/img.shape[0],4)
        h_scale=np.round(resized_image.shape[1]/img.shape[1],4)
        ss_box=[0,0,0,0]
        ss_box[0]=int(box2[0]*w_scale)
        ss_box[1]=int(box2[1]*h_scale)
        ss_box[2]=int(box2[2]*w_scale)
        ss_box[3]=int(box2[3]*h_scale)
        if(cur_iou>0.3):
            positive_box.append(ss_box)
        else:
            negative_box.append(ss_box)
    #write positive data
    positive = Element("positive")
    image_num.append(positive)
    for j in range(len(positive_box)):
        cnt=Element("a"+str(j))
        cnt.text = str(positive_box[j])
        positive.append(cnt)
        
    #write negative data
    negative = Element("negative")
    image_num.append(negative)
    for j in range(len(negative_box)):
        cnt=Element("a"+str(j))
        cnt.text = str(negative_box[j])
        negative.append(cnt)

    print("image : ",i,"nagetive : ",len(negative_box),"positive : ",len(positive_box),"total box : ",cnt_box)
    
    
    

indent(note)
```
